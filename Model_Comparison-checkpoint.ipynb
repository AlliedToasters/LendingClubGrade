```python
loans.requested_amnt = loans.requested_amnt.astype('int')*100 #Convert dollar amounts: floating point to int in cents
loans.loan_size = loans.loan_size.astype('int')*100
loans.monthly_pymnt = loans.monthly_pymnt.astype('int')*100
loans.income = loans.income.astype('int')*100
loans.revolving_credit = loans.revolving_credit.astype('int')*100
loans.prncp_paid = loans.prncp_paid.astype('int')*100
loans.total_paid = loans.total_paid.astype('int')*100

loans['APR'] = pd.to_numeric(loans.APR.str.slice(0, -1)) #Convert percentages (strings) to numeric values
loans['credit_utilization'] = pd.to_numeric(loans.credit_utilization.str.slice(0, -1))

loans.loan_term = pd.to_numeric(loans.loan_term.str.slice(0, 3)).astype('int') #Convert loan term to number
```

```python
#Convert employee length strings to numeric values

def get_emp_length(emp_length):
    """Takes our emp_length data (string) and returns numeric value
    in number of years.
    """
    unique_values = {         #keys taken from loans.emp_length.unique()
        '10+ years' : 10, 
        '< 1 year' : .5,      #between 0 and 1
        '3 years' : 3, 
        '9 years' : 9, 
        '4 years' : 4, 
        '5 years' : 5,
        '1 year' : 1, 
        '6 years' : 6, 
        '2 years' : 2, 
        '7 years' : 7, 
        '8 years' : 8, 
        'n/a': 0
    }
    return unique_values[emp_length]

loans.emp_length = loans.emp_length.apply(get_emp_length)
```

```python
#Want to convert dates into numeric format.

def convert_date(input_string):
    """changes date from format 'Mon-YYYY' to
    an integer number of months before Jan 1, 2016
    (when this data was published)
    """
    months_num = {
        'Jan' : 1,
        'Feb' : 2,
        'Mar' : 3,
        'Apr' : 4,
        'May' : 5,
        'Jun' : 6,
        'Jul' : 7,
        'Aug' : 8,
        'Sep' : 9,
        'Oct' : 10,
        'Nov' : 11,
        'Dec' : 12
    }
    mon = input_string[:3]
    year = int(input_string[-4:])
    if mon in months_num:
        num = months_num[mon]
    else:
        raise ValueError('{} not found in dictionary'.format(mon))
    months_passed = (12 - num) + ((2015 - year) * 12)
    if not str(months_passed).isnumeric():
        raise Exception('Error: return object not numeric: {}'.format(months_passed))
    return months_passed

loans.issue_date = loans.issue_date.apply(convert_date)
loans.first_cr_line = loans.first_cr_line.apply(convert_date)
```

```python
#Calculate number of months first credit line before loan issued
loans['fcl_before_loan'] = loans.first_cr_line - loans.issue_date
```

```python
#Convert home status to a numeric value

def convert_home(home_type):
    """Convert housing status to numeric value.
    Tries to rank numbers roughly by wealth-association with each
    homeownership status.
    """
    unique_values = {   #keys taken from loans.home_status.unique() 
        'OTHER' : 0, 
        'NONE' : 0,     #These three vague categories all assigned to zero
        'ANY' : 0,
        'RENT' : 1,
        'MORTGAGE' : 2, 
        'OWN' : 3
    }
    return unique_values[home_type]

loans.home_status = loans.home_status.apply(convert_home)
```

```python
# Convert grade to numeric value

def convert_grade(grade):
    """Converts borrower grade to numeric value"""
    unique_grades = { #keys taken from loans.grade.unique()
        'A1' : 35,
        'A2' : 34,
        'A3' : 33,
        'A4' : 32,
        'A5' : 31,
        'B1' : 30,
        'B2' : 29,
        'B3' : 28,
        'B4' : 27,
        'B5' : 26,
        'C1' : 25,
        'C2' : 24,
        'C3' : 23,
        'C4' : 22,
        'C5' : 21,
        'D1' : 20,
        'D2' : 19,
        'D3' : 18,
        'D4' : 17,
        'D5' : 16,
        'E1' : 15,
        'E2' : 14,
        'E3' : 13,
        'E4' : 12,
        'E5' : 11,
        'F1' : 10,
        'F2' : 9,
        'F3' : 8,
        'F4' : 7,
        'F5' : 6,
        'G1' : 5,
        'G2' : 4,
        'G3' : 3,
        'G4' : 2,
        'G5' : 1
    }
    return unique_grades[grade]

loans.grade = loans.grade.apply(convert_grade)
```

```python
#This is a custom-build cross validation function I wrote for a previous assignment
#I couldn't find a built-in cv method for the statsmodels regression.

from sklearn.metrics import r2_score
def cross_validate_ols(data, train, target, folds=10):
    """Takes a pd.DataFrame, list of training variables, and a list of target variables.
    Optional keyword argument folds specifies desired number of folds/holdout groups.
    Does cross validation test with n number of folds and returns a list of Rsquared values
    equal in length to the number of folds. Uses smf.ols for regression statistics.
    """
    n = int(folds) #n tracks desired number of folds
    folds = dict() #will track selected indices
    inds = pd.Series(data.index) #use series type for .sample() method
    selected = [] #will track indices already selected
    mod = len(data)%n #handle remainder of len/folds
    group_size = int((len(data)-mod)/n) #ensures proper integer group size
    for fold in range(1, n+1):  #This loop selects random samples for each group
        if fold == n:
            group_size += mod #add remainder to last group
        sample = inds[~inds.index.isin(selected)].sample(group_size) #excluded indices already selected
        selected += list(sample)
        name = 'sample_{}'.format(fold)
        folds[name] = sample
    scores = [] #This will be output
    for sample in folds:
        learn = data[~data.index.isin(folds[sample])]
        test_set = data[data.index.isin(folds[sample])]
        formula = target + ' ~ '
        for feature in train:
            formula += feature + '+'
        formula = formula[:-1]
        reg = smf.ols(formula=formula, data=learn).fit() #fit model
        Y_ = reg.predict(test_set[train])
        scores.append(r2_score(test_set[target], Y_))
    return scores
```

```python
scores = cross_validate_ols(loans, in_var, 'grade', folds=10)
scores = pd.Series(scores)
print(scores.mean(), scores.std())
#We can see that this model performs fairly consistently.
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

```python
def ranked_weights(dist): #This is my simple rank-based weighting function
    """Returns an array with same dim as dist,
    exploits the fact that dist is ordered by
    rank of distance. Uses index to rank distances.
    """
    lgth, wdth = dist.shape
    wgts = np.zeros(shape=dist.shape)
    for d in range(lgth):
        for i in range(wdth):
            wgts[d][i] = wdth - i 
    return wgts

def test_Ks(X, Y, k_min=1, k_range=100, step=1, weight='uniform', folds=10):
    cols = ['k', 'weight', 'mean_score', 'score_std']
    results = pd.DataFrame(columns=cols)
    i = 1
    for k in range(k_min, k_range, step):
        mdl = KNeighborsRegressor(n_neighbors=k, weights=weight)
        score = cross_val_score(mdl, X, Y, cv=folds)
        new_results = pd.DataFrame(columns=cols)
        new_results.set_value(i, 'k', k)
        new_results.set_value(i, 'weight', weight)
        new_results.set_value(i, 'mean_score', score.mean())
        new_results.set_value(i, 'score_std', score.std())
        i += 1
        results = pd.concat([results, new_results])
    results.upper_bound = results.mean_score + results.score_std
    results.lower_bound = results.mean_score - results.score_std
    return results
```

```python
knn = KNeighborsRegressor(n_neighbors=200, weights=ranked_weights)

knn.fit(X, Y)
Y_ = knn.predict(X)

pred_v_act = pd.DataFrame()
pred_v_act['Y'] = Y
pred_v_act['Y_'] = Y_

sns.jointplot(x='Y', y='Y_', data=pred_v_act);
```
